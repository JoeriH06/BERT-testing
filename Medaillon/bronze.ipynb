#  This is the bronze layer where we gather the data.  

In our case we have pre-made datasets, these datasets where made by AI.  
The datasets that where created are in blob-format, with the ID, Title, and content sepperated.  


```python
# First we will import the libraries.
import pandas as pd
import os
import json
```

```python
# Then we will read the data from the Data folder. These JSON files are untouched and raw data.
# we will read the data into a pandas dataframe for easier manipulation and analysis.

bronze_data = pd.DataFrame()

for file in os.listdir('../Data/raw'):
    if file.endswith('01.json'):
        with open(os.path.join('../Data/raw', file), 'r') as f:
            data = json.load(f)
            bronze_data = pd.concat([bronze_data, pd.DataFrame([data])], ignore_index=True)

# we check if the data is loaded correctly by displaying the first few rows of the dataframe.
bronze_data.head(10)
```

```python
# Now that we put the bronze data into a dataframe, we can go on to the silver layer.
# This silver layer needs to bronze dataframe so we have to pass this through to the next notebook.

bronze_data.to_csv('../Data/bronze/bronze_data.csv', index=False)
```
