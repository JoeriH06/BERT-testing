In our case we have pre-made datasets, these datasets where made by AI.  
The datasets that where created are in blob-format, with the ID, Title, and content sepperated.  


```python
# First we will import the libraries.
import pandas as pd
import os
import json
import requests
```

```python
# Path to uploaded document
file_path = '../Data/raw/doc_01.txt'

# Extract document id from filename
document_id = os.path.splitext(os.path.basename(file_path))[0]

# Read raw text
with open(file_path, 'r', encoding='utf-8') as f:
    text = f.read()

# Store raw data in dataframe
bronze_data = pd.DataFrame([{
    "document_id": document_id,
    "raw_text": text
}])

# Save bronze snapshot
output_path = f'../Data/bronze/{document_id}_bronze.json'
bronze_data.to_json(output_path, orient='records')

bronze_data
```
